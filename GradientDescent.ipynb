# 예측값과 변화율

y = ax + b
- a는 보통 가중치를 의미하는 w나 계수를 의미하는 θ로 표기한다.
- y는 Ŷ으로 표기한다.
위와 같은 일련의 규칙으로 수식을 다시 표현하면 아래와 같이 표기할 수 있다.
Ŷ = wx + b
- w와 b는 알고리즘이 찾은 규칙을 의미하고 Ŷ은 <b>예측값</b>을 의미한다.

예측값
<br>
- 입력과 타깃(출력) 데이터, 즉 x,y를 통해 규칙(w,b)를 발견하면 모델을 만들었다고 표현한다.
- 그 모델에 새로운 입력값(x)을 넣으면 출력(Ŷ)이 나오는데 이 값이 모델을 통해 예측한 값이 된다.
- y와 Ŷ의 구분은 타깃데이터인지 예측값인지 구분짓기 위함이다.

예측값으로 올바른 모델 찾기
<ol>
<li>무작위로 w와 b를 정한다.</li>
<li>x에서 샘플 하나를 선택하여 Ŷ을 계사한다.</li>
<li>Ŷ과 선택한 샘플의 진짜 y를 비교한다.</li>
<li>Ŷ이 y와 가까와지도록 w,b를 조정한다.</li>
<li>모든 샘플을 처리할 때까지 다시 2 ~ 4를 반복한다.</li>
</ol>
